```python
import pandas as pd

# üìå Load both CSV files
train_df = pd.read_csv('CBIS-DDSM/csv/train.csv')
test_df = pd.read_csv('CBIS-DDSM/csv/test.csv')

# ‚úÖ Append test to the end of train
combined_df = pd.concat([train_df, test_df], ignore_index=True)

# üìÅ Save combined file
combined_path = 'CBIS-DDSM/csv/train_plus_test.csv'
combined_df.to_csv(combined_path, index=False)

print(f"‚úÖ Combined dataset saved as: {combined_path}")
print(f"üìä Total rows: {len(combined_df)} (Train: {len(train_df)}, Test: {len(test_df)})")
```
