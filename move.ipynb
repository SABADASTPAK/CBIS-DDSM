```python
import os
import shutil
from tqdm.notebook import tqdm

# üìÅ Define source and target folders
image_folders = [
    'CBIS-DDSM/NEW/images',
    'CBIS-DDSM/NEW/images2'
]
label_folders = [
    'CBIS-DDSM/NEW/labels',
    'CBIS-DDSM/NEW/labels2'
]

target_image_folder = 'CBIS-DDSM/NEW/IMAGES'
target_label_folder = 'CBIS-DDSM/NEW/LABELS'

# üìå Create target folders if not exist
os.makedirs(target_image_folder, exist_ok=True)
os.makedirs(target_label_folder, exist_ok=True)

# üîÅ Move images
print("üì¶ Moving image files...")
for folder in image_folders:
    for filename in tqdm(os.listdir(folder), desc=f"Merging {os.path.basename(folder)}"):
        src = os.path.join(folder, filename)
        dst = os.path.join(target_image_folder, filename)

        try:
            if os.path.isfile(src):
                shutil.move(src, dst)
        except Exception as e:
            print(f"‚ùå Failed to move {filename}: {e}")

# üîÅ Move labels
print("\nüì¶ Moving label files...")
for folder in label_folders:
    for filename in tqdm(os.listdir(folder), desc=f"Merging {os.path.basename(folder)}"):
        src = os.path.join(folder, filename)
        dst = os.path.join(target_label_folder, filename)

        try:
            if os.path.isfile(src):
                shutil.move(src, dst)
        except Exception as e:
            print(f"‚ùå Failed to move {filename}: {e}")

print("\n‚úÖ All files moved to IMAGES/ and LABELS/")
```
