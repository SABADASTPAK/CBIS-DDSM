```python
import pandas as pd

csv_path = 'CBIS-DDSM/csv/mass_train_jpg2_bbox.csv'
df = pd.read_csv(csv_path)

df['new name'] = ''

# Group by patient_id, left or right breast, and image view
grouped = df.groupby(['patient_id', 'left or right breast', 'image view'])

for (patient_id, breast_side, view), group in grouped:
    new_name = f"{patient_id}_{breast_side}_{view}".replace(" ", "_")
    df.loc[group.index, 'new name'] = new_name

output_path = 'CBIS-DDSM/csv/train.csv'
df.to_csv(output_path, index=False)

print("✅ Done! 'train.csv' has been saved in your Google Drive with the 'new name' column.")
```
