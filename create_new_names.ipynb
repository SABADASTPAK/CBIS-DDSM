```python
import pandas as pd

# üîΩ Step 1: Load CSV file
csv_path = 'CBIS-DDSM/csv/mass_train_jpg2_bbox.csv'  # Update path if needed
df = pd.read_csv(csv_path)

# üîÑ Step 2: Create 'new names' column by concatenating specific columns
df['new names'] = (
    df['patient_id'].astype(str) + '_' +
    df['left or right breast'].astype(str).str.upper() + '_' +
    df['image view'].astype(str).str.upper() + '_' +
    df['abnormality id'].astype(str)
)

# ‚úÖ Step 3: Display result (optional)
print(df[['patient_id', 'left or right breast', 'image view', 'abnormality id', 'new names']].head())

# üíæ Step 4: Save new DataFrame to a CSV (optional)
df.to_csv('CBIS-DDSM/csv/mass_train_jpg2_bbox_with_newnames.csv', index=False)
```

```python
# üîç Step: Check for duplicate 'new names'
duplicates = df[df['new names'].duplicated(keep=False)]

# üìä Display duplicates (if any)
if not duplicates.empty:
    print("‚ö†Ô∏è Duplicate 'new names' found:")
    print(duplicates[['patient_id', 'left or right breast', 'image view', 'abnormality id', 'new names']])
else:
    print("‚úÖ No duplicate 'new names' found.")
```
