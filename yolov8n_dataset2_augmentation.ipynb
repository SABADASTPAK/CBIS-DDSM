```python
from ultralytics import YOLO

# Step 1: Set paths and model
data_yaml = '/content/drive/MyDrive/yolo_dataset2/data.yaml'
model_arch = 'yolov8n.pt'  # lightweight, good for quick experiments

# Step 2: Load model
model = YOLO(model_arch)

# Step 3: Train with configuration
model.train(
    data=data_yaml,
    epochs=100,
    imgsz=640,
    batch=16,
    project='/content/drive/MyDrive/yolo_results2',
    name='yolov8n_dataset2_augmentation',
    exist_ok=True,

    # âœ… Augmentations (YOLOv8 has built-in support)
    degrees=10,           # Random rotation
    scale=0.5,            # Scale image
    shear=0.0,            # No shearing by default
    perspective=0.0,      # No perspective distortion
    flipud=0.0,           # Vertical flip probability
    fliplr=0.5,           # Horizontal flip probability (common in medical imaging)
    mosaic=1.0,           # Mosaic augmentation
    mixup=0.2,            # Mixup augmentation
    hsv_h=0.015,          # Hue shift
    hsv_s=0.7,            # Saturation shift
    hsv_v=0.4,            # Brightness shift

    # ğŸ”§ Training improvements
    patience=20,          # Early stopping if no improvement
    val=True,             # Run validation every epoch
    device=0,              # GPU index or 'cpu'
    save_period=5  # â† save weights every 5 epochs
)
```

```python
!pip install ultralytics

import matplotlib.pyplot as plt
import cv2
import torch
import yaml
from ultralytics.data import YOLODataset

# Path to your dataset yaml
data_yaml = "/content/drive/MyDrive/yolo_dataset2/data.yaml"

# Load yaml as dict
with open(data_yaml, "r") as f:
    data_dict = yaml.safe_load(f)

# Create dataset with augmentations
dataset = YOLODataset(data=data_dict, img_size=640, augment=True)

# Show a few augmented samples
for i in range(3):  # visualize 3 samples
    sample = dataset[i]
    img = sample["img"].permute(1, 2, 0).numpy()  # (H, W, C)
    img = (img * 255).astype("uint8")

    plt.figure(figsize=(8, 8))
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"Augmented Sample {i+1}")
    plt.show()
```
